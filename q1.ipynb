(1) $ x'(-1)$を求めよ。

(2)
\begin{equation} \lim_{f \to 1} \frac{x(f)+x(-f)-2}{f-1} \end{equation}
を求めよ。

\begin{equation} -x'(-f)=x'(f)+2 \end{equation}.

\begin{equation} x'(-1)=-3 \end{equation}
